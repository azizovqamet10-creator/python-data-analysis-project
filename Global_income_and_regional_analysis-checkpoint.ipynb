                                                                
                                                         Data Visualization
                   In this section, I visualize the distribution of income groups across different geographical regions.
                                                                

```python
viz_data = pd.crosstab(df_copy['Region'], df_copy['IncomeGroup'])

viz_data.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Income Groups by Region')
plt.xlabel('Regions')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45) 
plt.legend(title='Income Group', bbox_to_anchor=(1.0, 1.0))
plt.show()
```

```python
income_counts = df_copy['IncomeGroup'].value_counts()


plt.figure(figsize=(8, 8))


income_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)


plt.title('Global Distribution of Income Groups', fontsize=12, fontweight='bold')
plt.ylabel('') 

plt.show()
```
